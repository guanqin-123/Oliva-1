```python
def solved(df):
    
    return df[df.bab_baseline_status!= 'Status.UNKNOWN'].shape[0], df[df.abcrown_status!= 'Status.UNKNOWN'].shape[0], df[df.neuralsat_status!= 'Status.UNKNOWN'].shape[0], df[df.oliva_greedy_status!= 'Status.UNKNOWN'].shape[0], df[df.oliva_balance_status!= 'Status.UNKNOWN'].shape[0]

def timeAverage(df):
    df = df[(df.oliva_greedy_status!= 'Status.UNKNOWN')|(df.oliva_balance_status!= 'Status.UNKNOWN')|(df.abcrown_status!= 'Status.UNKNOWN')|(df.neuralsat_status!= 'Status.UNKNOWN')]
    return (
        round(df.bab_baseline_time.mean(), 2),
        round(df.abcrown_time.mean(), 2),
        round(df.neuralsat_time.mean(), 2),
        round(df.oliva_greedy_time.mean(), 2),
        round(df.oliva_balance_time.mean(), 2)
    )
    
def print_lines(columName, all_data):
    a, b, c, d , e = solved(all_data)
    at, bt, ct, dt, et = timeAverage(all_data)
    print(f'{columName} &{a}&{at} &{b}&{bt} &{c}&{ct} &{d}&{dt} &{e}&{et}  \\\\')
    
```

```python
def getTable1():
    all_data1 = pd.read_csv('all_results/all_mnistL2.csv')
    print_lines( '$\mnist_{{{\ltwo}}}$ ', all_data1)
    
    all_data2 = pd.read_csv('all_results/all_mnistL4.csv')
    print_lines( '$\mnist_{{{\lfour}}}$ ', all_data2)
    
    all_data3 = pd.read_csv('all_results/all_oval21_base.csv')
    print_lines( '$\OVAL_{{\\base}}$', all_data3)
    
    all_data4 = pd.read_csv('all_results/all_oval21_deep.csv')
    print_lines( '$\OVAL_{{\\deep}}$', all_data4)
    
    all_data5 = pd.read_csv('all_results/all_oval21_wide.csv')
    print_lines( '$\OVAL_{{\wide}}$', all_data5)
```

```python
def calculate_ratio_statistics(df, data_name):
    tools = {
        'Oliva$^{GR}$': ('oliva_greedy_time', 'oliva_greedy_status'),
        'Oliva$^{SA}$': ('oliva_balance_time', 'oliva_balance_status')
    }
    
    stats_data = {}  # Dictionary to store statistics for each method
    
    for tool_name, (time_col, status_col) in tools.items():
        # Filter out UNKNOWN status cases for each method
        valid_cases = df[status_col] != "Status.UNKNOWN"
        filtered_df = df[valid_cases]
       
        ratio = filtered_df['bab_baseline_time'] / filtered_df[time_col]
        mask = ratio <= 150
        filtered_ratio = ratio[mask]
        
        # Calculate statistics
        stats_data[tool_name] = {
            'min': filtered_ratio.min(),
            'max': filtered_ratio.max(),
            'median': filtered_ratio.median(),
            'mean': filtered_ratio.mean()
        }
    
    # Print statisticsxx
    print(f"\nStatistics for {data_name}:")
    for tool_name, stats in stats_data.items():
        print(f"\n{tool_name}:")
        print(f"Min: {stats['min']:.2f}")
        print(f"Max: {stats['max']:.2f}")
        print(f"Median: {stats['median']:.2f}")
        print(f"Mean: {stats['mean']:.2f}")
    
    return stats_data
stats = calculate_ratio_statistics(combined_df, "overall")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.colors as mcolors
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas

# Assuming any_known_df is already defined
methods = ['sa0', 'sa1', 'sa2', 'sa3', 'sa4', 'sa5']
method_names = ['sa0', 'sa1', 'sa2', 'sa3', 'sa4', 'sa5']

# Calculate ratios
scatter = combined_df[["id", "eps", "dfs_time"] + [f"{method}_time" for method in methods]]
for method in methods:
    scatter[f'ratio_{method}'] = scatter['dfs_time'] / scatter[f'{method}_time']

# Create a list to store data for each method
plot_data = []

for method in methods:
    df = pd.DataFrame({
        'Time': scatter[f'{method}_time'],
        'Ratio': scatter[f'ratio_{method}'],
        'Method': method_names[methods.index(method)]
    })
    plot_data.append(df)

# Combine all data
plot_data = pd.concat(plot_data, ignore_index=True)

# Filter out data points above 1000s
plot_data = plot_data[plot_data['Time'] <= 1000]

# Define ratio labels - removed 10+ category
ratio_labels = ['0-0.5', '0.5-1', '1-2', '2-5', '5-10']

# Define 6 time bins
time_bins = [0, 100, 200, 400, 600, 800, 1000]
time_labels = ['0-100s', '100-200s', '200-400s', '400-600s', '600-800s', '800-1000s']
plot_data['Time_bin'] = pd.cut(plot_data['Time'], bins=time_bins, labels=time_labels, include_lowest=True, right=False)

# Create a custom colormap
light_gray = '#f0f0f0'  # Light gray color
colors = plt.cm.Blues(np.linspace(0.2, 1, 256))  # Using Blues colormap
custom_cmap = mcolors.LinearSegmentedColormap.from_list("custom", [light_gray] + list(colors))
custom_cmap.set_under(light_gray)  # Set color for values below vmin

def bin_ratios(ratios, labels):
    # Modified bins to remove the 10+ category
    bins = [0, 0.5, 1, 2, 5, 10]
    return pd.cut(ratios, bins=bins, labels=labels, include_lowest=True, right=False)

# Find the global maximum for consistent color scaling
global_max = 0
for method in method_names:
    method_data = plot_data[plot_data['Method'] == method]
    heatmap_data = pd.DataFrame(index=ratio_labels, columns=time_labels)
    for time_bin in time_labels:
        time_data = method_data[method_data['Time_bin'] == time_bin]
        binned_ratios = bin_ratios(time_data['Ratio'], ratio_labels)
        counts = binned_ratios.value_counts().reindex(ratio_labels).fillna(0)
        heatmap_data[time_bin] = counts
    global_max = max(global_max, heatmap_data.max().max())

def create_and_save_subplot(method, method_name, index):
    fig, ax = plt.subplots(figsize=(8, 5))  # Reduced subplot size
    
    # Create a pivot table for the heatmap
    method_data = plot_data[plot_data['Method'] == method_name]
    heatmap_data = pd.DataFrame(index=ratio_labels, columns=time_labels)
    
    for time_bin in time_labels:
        time_data = method_data[method_data['Time_bin'] == time_bin]
        binned_ratios = bin_ratios(time_data['Ratio'], ratio_labels)
        counts = binned_ratios.value_counts().reindex(ratio_labels).fillna(0)
        heatmap_data[time_bin] = counts
    
    # Create heatmap with the custom colormap and consistent vmax
    sns.heatmap(heatmap_data, ax=ax, cmap=custom_cmap, 
                cbar=False, annot=False, fmt='d', linewidths=0.5, linecolor='white', square=True,
                vmin=0.1, vmax=global_max)
    
    # Annotate heatmap
    annotate_heatmap(ax, heatmap_data)
    
    # Set labels
    ax.set_xlabel('Time', fontsize=16, fontweight='bold')  # Increased font size
    
    # Set x-axis labels
    ax.set_xticklabels(time_labels, rotation=45, ha='right', fontsize=14)  # Increased font size
    
    # Set y-axis labels
    ax.set_yticks(np.arange(len(ratio_labels)) + 0.5)
    ax.set_yticklabels(ratio_labels, fontsize=14)  # Increased font size
    
    # Add reference lines - adjusted positions for 5 categories instead of 6
    ax.axhline(y=2, color='black', linestyle='--', linewidth=2)  # Line between 1 and 2
    # ax.axhline(y=3.5, color='white', linestyle='--', linewidth=2)  # Line between 2-5 and 5-10
    
    # Add text labels for ranges - adjusted positions
    ax.text(-0.5, 0.8, 'Slower', rotation=90, va='center', ha='right', fontsize=14, fontweight='bold')
    ax.text(-0.5, 2.5, 'Faster', rotation=90, va='center', ha='right', fontsize=14, fontweight='bold')
    ax.text(-0.5, 4.2, 'Much\nFaster', rotation=90, va='center', ha='right', fontsize=14, fontweight='bold')

    # Adjust layout
    plt.tight_layout()
    plt.subplots_adjust(left=0.15, right=1)

    # Save the subplot
    plt.savefig(f'subplot_{index}_{method_name}.png', format='png', dpi=300, bbox_inches='tight')
    plt.close(fig)

def annotate_heatmap(ax, data):
    for i, row in enumerate(data.index):
        for j, col in enumerate(data.columns):
            value = data.iloc[i, j]
            if pd.notna(value):
                text_color = 'white' if value > data.max().max() / 2 else 'black'
                ax.text(j + 0.5, i + 0.5, f'{int(value)}', 
                        ha='center', va='center', color=text_color, fontsize=16, fontweight='bold')  # Increased font size

# Create and save individual subplots
for i, (method, method_name) in enumerate(zip(methods, method_names)):
    create_and_save_subplot(method, method_name, i)

# Create the main figure with 2x3 layout
fig, axes = plt.subplots(2, 3, figsize=(30, 16))  # Reduced overall figure size
fig.suptitle('Heatmap of Time vs Ratio for Each Method', fontsize=28, fontweight='bold', y=1.02)

# Populate the main figure
for i, (method, method_name) in enumerate(zip(methods[:6], method_names[:6])):  # Only first 6 methods
    row, col = divmod(i, 3)  # Arrange in 2x3 grid
    
    # Load the saved subplot image
    img = plt.imread(f'subplot_{i}_{method_name}.png')
    axes[row, col].imshow(img)
    axes[row, col].axis('off')

# Adjust layout with reduced horizontal spacing and increased vertical spacing
plt.tight_layout()
plt.subplots_adjust(top=0.95, bottom=0.05, left=0.05, right=0.98, hspace=0.4, wspace=0.05)

plt.savefig('Fig7sigma_comprehensive_time_ratio_heatmap.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```
